```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import time

img = cv.imread('lena512color.tiff')
img2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)

start = time.time()

# Convert image to grayscale
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Display grayscale image
plt.figure()
plt.imshow(gray_img, cmap='gray')
plt.show()

print("run time: " + str(time.time() - start))

# Save RGB data as CSV format with delimiter=""
np.savetxt(fname='blue_peppers.csv', delimiter=",", X=img2[:, :, 2].reshape(-1, 1))
np.savetxt(fname='green_peppers.csv', delimiter=",", X=img2[:, :, 1].reshape(-1, 1))
np.savetxt(fname='red_peppers.csv', delimiter=",", X=img2[:, :, 0].reshape(-1, 1))
```

```python
%%cuda
#include <stdio.h>
#include <stdlib.h>
#include <cuda.h>
#include "cuda_runtime.h"
#include "device_launch_parameters.h"
#include <chrono>
#include <math.h>

__global__ void ctg(unsigned char *red, unsigned char *blue, unsigned char *green, unsigned char *grayImage, int width, int height) {
    int Col = threadIdx.x + blockIdx.x * blockDim.x;
    int Row = threadIdx.y + blockIdx.y * blockDim.y;

    if (Col < width && Row < height) {
        int grayOffset = Row * width + Col;
        unsigned char r = red[grayOffset];
        unsigned char g = green[grayOffset];
        unsigned char b = blue[grayOffset];
        grayImage[grayOffset] = 0.21f * r + 0.71f * g + 0.07f * b;
    }
}

__global__ void edgeDetector(unsigned char *input, unsigned char *output, const int height, const int width) {
    int col = threadIdx.x + blockIdx.x * blockDim.x;
    int row = threadIdx.y + blockIdx.y * blockDim.y;
    float filterX[3][3] = { {-1, 0, 1}, {2, 0, 2}, {-1, 0, 1} };
    float filterY[3][3] = { {1, 2, 1}, {0, 0, 0}, {-1, -2, -1} };
    if (col < width && row < height) {
        int offset = row * width + col;
        float outputX = 0, outputY = 0;
        for (int convolutionRow = -1; convolutionRow <= 1; convolutionRow++) {
            for (int convolutionCol = -1; convolutionCol <= 1; convolutionCol++) {
                int curRow = row + convolutionRow;
                int curCol = col + convolutionCol;
                if (curCol < width && curRow < height && curCol > -1 && curRow > -1) {
                    int curOffset = curRow * width + curCol;
                    outputX += input[curOffset] * filterX[convolutionRow + 1][convolutionCol + 1];
                    outputY += input[curOffset] * filterY[convolutionRow + 1][convolutionCol + 1];
                }
            }
        }
        output[offset] = sqrt(outputX * outputX + outputY * outputY);
    }
}

__global__ void gauss(unsigned char* Pin, unsigned char* Pout, int width, int height)
	{
	float filter[5][5] = { {2,4,5,4,2}, {4,9,12,9,4},{5,12,15,12,5},{4,9,12,9,4},{2,4,5,4,2} };
	for (int i = 0; i < 5; i++) for (int j = 0; j < 5; j++) {
		filter[i][j] = filter[i][j] / 159;
		}
	int col = threadIdx.x + blockIdx.x * blockDim.x;
	int row = threadIdx.y + blockIdx.y * blockDim.y;
	int n = 2;

	float Pvalue = 0;
	int Pnumber = 0;
	for (int mi = -n; mi <= n; mi++) {
		for (int mj = -n; mj <= n; mj++) {
			if (row + mi >= 0 && row + mi < width && col + mj >= 0 && col + mj < height) {
				Pvalue += Pin[((row + mi) * width + (col + mj))] * filter[mi + n][mj + n];
				Pnumber++;
				}
			}
		}
	Pout[(row * width + col)] = Pvalue;
	}

  __global__ void laplus(unsigned char* Pin, unsigned char* Pout, int width, int height)
	{
	float filter[9][9] = {
			{0,0,3,2,2,2,3,0,0},
			{0,2,3,5,5,5,3,2},
			{3,3,5,3,0,3,5,3,3},
			{2,5,3,-12,-23,-12,3,5,2},
			{2,5,3,-12,-23,-12,-3,-5,-2},
			{3,3,5,3,0,3,5,3,3},
			{0,2,3,5,5,5,3,2,0},
			{0,0,3,2,2,3,0,0} };
	int col = threadIdx.x + blockIdx.x * blockDim.x;
	int row = threadIdx.y + blockIdx.y * blockDim.y;
	int n = 4;

	float Pvalue = 0;
	int Pnumber = 0;
	for (int mi = -n; mi <= n; mi++) {
		for (int mj = -n; mj <= n; mj++) {
			if (row + mi >= 0 && row + mi < width && col + mj >= 0 && col + mj < height) {
				Pvalue += Pin[((row + mi) * width + (col + mj))] * filter[mi + n][mj + n];
				Pnumber++;
				}
			}
		}
	Pout[(row * width + col)] = Pvalue / Pnumber;
	}


int main() {
    FILE *fp;
    char blue[512*512];
    unsigned char *d_blue;

    char red[512*512];
    unsigned char *d_red;

    char green[512*512];
    unsigned char *d_green;

    unsigned char grayImage[512*512];
    unsigned char *d_grayImage;

    unsigned char edgeImage[512*512];
    unsigned char *d_edgeImage;

    unsigned char gaussImage[512*512];
    unsigned char *d_gaussImage;

    unsigned char laplusImage[512*512];
    unsigned char *d_laplusImage;

    const int height = 512;
    const int width = 512;
    const int PCount = height * width;
    const int MemoryNeed = PCount * sizeof(unsigned char);

    cudaMalloc((void**)&d_blue, MemoryNeed);
    cudaMalloc((void**)&d_green, MemoryNeed);
    cudaMalloc((void**)&d_red, MemoryNeed);
    cudaMalloc((void**)&d_grayImage, MemoryNeed);
    cudaMalloc((void**)&d_edgeImage, MemoryNeed);
    cudaMalloc((void**)&d_gaussImage, MemoryNeed);
    cudaMalloc((void**)&d_laplusImage, MemoryNeed);


    fp = fopen("blue_peppers.csv", "r");
    fread(blue, sizeof(char), PCount, fp);
    fclose(fp);

    fp = fopen("green_peppers.csv", "r");
    fread(green, sizeof(char), PCount, fp);
    fclose(fp);

    fp = fopen("red_peppers.csv", "r");
    fread(red, sizeof(char), PCount, fp);
    fclose(fp);

    cudaMemcpy(d_blue, blue, MemoryNeed, cudaMemcpyHostToDevice);
    cudaMemcpy(d_green, green, MemoryNeed, cudaMemcpyHostToDevice);
    cudaMemcpy(d_red, red, MemoryNeed, cudaMemcpyHostToDevice);

    dim3 dimGrid(512 / 32 + 1, 512 / 32 + 1, 1);
    dim3 dimBlock(32, 32, 1);

    auto start = std::chrono::high_resolution_clock::now();
    ctg<<<dimGrid, dimBlock>>>(d_red, d_blue, d_green, d_grayImage, height, width);
    cudaDeviceSynchronize();

    edgeDetector<<<dimGrid, dimBlock>>>(d_grayImage, d_edgeImage, height, width);
    cudaDeviceSynchronize();

    gauss<<<dimGrid, dimBlock>>>(d_grayImage, d_gaussImage, height, width);
    cudaDeviceSynchronize();

    gauss<<<dimGrid, dimBlock>>>(d_grayImage, d_laplusImage, height, width);
    cudaDeviceSynchronize();

    auto end = std::chrono::high_resolution_clock::now();
    double time_taken = std::chrono::duration_cast<std::chrono::nanoseconds>(end - start).count();
    time_taken *= 1e-9;
    printf("Time taken: %lf seconds\n", time_taken);

    cudaMemcpy(grayImage, d_grayImage, MemoryNeed, cudaMemcpyDeviceToHost);
    cudaMemcpy(edgeImage, d_edgeImage, MemoryNeed, cudaMemcpyDeviceToHost);
    cudaMemcpy(gaussImage, d_gaussImage, MemoryNeed, cudaMemcpyDeviceToHost);
    cudaMemcpy(laplusImage, d_laplusImage, MemoryNeed, cudaMemcpyDeviceToHost);



    return 0;
}
```
